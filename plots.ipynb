```python
rewards_train = []
with open('data/1577332521.703821_rewards_train.txt', 'r') as f:
    for line in f:
        rewards_train.append(float(line))
```

```python
len(rewards_train)
```

```python
max_rewards_train = []
with open('data/1577332521.703821_max_reward_train.txt', 'r') as f:
    for line in f:
        max_rewards_train.append(float(line))
len(max_rewards_train)
```

```python
steps_train = []
with open('data/1577332521.703821_steps_train.txt', 'r') as f:
    for line in f:
        steps_train.append(int(line))
len(steps_train)
```

```python
rewards_test = []
with open('data/1577332521.703821_rewards_test.txt', 'r') as f:
    for line in f:
        rewards_test.append(float(line))
len(rewards_test)
```

```python
tmp = []
i = 0
while i < len(rewards_test):
    if (i + 1) % 100 == 0:
        tmp.append(rewards_test[i] / 100)
    i += 1
```

```python
len(tmp)
```
